```python
import cv2
from matplotlib import pyplot as plt

def show_image(image, cmap=None, title="Image"):
    plt.figure(figsize=(5,5))
    if cmap:
        plt.imshow(image, cmap=cmap)
    else:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image with the correct name
img_color = cv2.imread('cube.jpg.jpg', 1)

if img_color is not None:
    show_image(img_color, title='Color Image')
    # Save the image with a new name
    file_name = 'saved_cube_colab.jpg'
    cv2.imwrite(file_name, img_color)
    print(f"Image saved as {file_name}")
else:
    print("Image not found or could not be loaded.")

```

```python
import cv2
import matplotlib.pyplot as plt

# Read image as RGB (OpenCV loads as BGR by default)
img = cv2.imread('cube.jpg.jpg')

# Convert BGR to RGB for correct color display in matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axis for a cleaner look
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# List all files in the Colab working directory
print("Files in working directory:", os.listdir())

# Use your image's exact filename
img_path = 'cube.jpg.jpg'

# Load the image
img = cv2.imread(img_path)
if img is None:
    print(f"Image not found or could not be loaded: {img_path}")
else:
    # Convert to YCrCb color space
    img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
    # Convert YCrCb to RGB for proper display in matplotlib
    img_ycrcb_rgb = cv2.cvtColor(img_ycrcb, cv2.COLOR_YCrCb2RGB)
    plt.imshow(img_ycrcb_rgb)
    plt.title('YCrCb Image (converted for display)')
    plt.axis('off')
    plt.show()
```
